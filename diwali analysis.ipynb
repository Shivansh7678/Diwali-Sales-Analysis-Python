**Exploratory Data Analysis
Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

**From above graph we can see most of the buyers are females as compared to mens

**Age analysis

```python
df.columns
```

```python
sns.countplot(data=df,x='Age Group')
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')#with hue which tells us about gender
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

Analysis according to age of customers 26-35 spend most of money

**State Analysis

```python
df.columns
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

we can see abovethat UP,Maharstra ,karnataka orders the most and spend high amount

**Occupation

```python
sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
```

From above analysis we can see most of the buyers are from IT sector as well as from Healthcare sector.

**Product category

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(27,10)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
sns.set(rc={'figure.figsize':(30,10)})
```

From above graph we see that most sold items are Food,Clothing and Electronics

```python
sales_Id=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=sales_Id)
sns.set(rc={'figure.figsize':(30,10)})
```

From above graph we see sales with product_id and orders.
